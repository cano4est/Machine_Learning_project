```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report
```

```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
iris.target
```

```python
data_plot = pd.DataFrame(iris.data,columns=iris.feature_names)
data_plot['iris'] = iris.target
```

```python
sns.pairplot(
    data_plot,
    hue = 'iris',
    palette = 'Dark2'
)
plt.show()
```

```python
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
gaussian = GaussianNB()
gaussian.fit(X_train, y_train)
```

```python
y_pred_train = gaussian.predict(X_train)
plt.figure(figsize=(4,4))
cm = confusion_matrix(y_train,y_pred_train)
sns.heatmap(cm, annot=True, annot_kws={"fontsize":18},cbar=False, fmt='.0f')
plt.show()
report = classification_report(y_train, y_pred_train, target_names=iris.target_names)
print('Classification Report:')
print(report)
```

```python
y_pred_test = gaussian.predict(X_test)
plt.figure(figsize=(4,4))
cm = confusion_matrix(y_test,y_pred_test)
sns.heatmap(cm, annot=True, annot_kws={"fontsize":18},cbar=False, fmt='.0f')
plt.show()
report = classification_report(y_test, y_pred_test, target_names=iris.target_names)
print('Classification Report:')
print(report)
```
